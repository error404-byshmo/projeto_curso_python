### Módulo 6: Funções (Reutilização de Código)

### 1. Definição de Funções (`def`)
Funções são blocos de código nomeados e reutilizáveis, definidos pela palavra-chave `def`. Elas tornam seu código modular e mais fácil de manter.

```python
# Exemplo: Função Simples
def saudar(nome):
    return f"Olá, {nome}. Bem-vindo de volta!"

mensagem = saudar("Aluno")
print(mensagem)
```

### 2. Funções Matemáticas (Exemplo Fatorial)
Funções são perfeitas para encapsular cálculos complexos ou lógicos (como o Fatorial, Média ou Delta).

```python
# Exemplo: Fatorial de um número (n!)
def fatorial(n: int) -> int:
    if n == 0:
        return 1
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

print(f"Fatorial de 4: {fatorial(4)}") # 4*3*2*1 = 24
print(f"Fatorial de 0: {fatorial(0)}") # 1
```

### 3. Exercícios Práticos
#### Exercício 1: Classificação de Notas
Crie uma função `classifica_nota(nota)` que retorna 'Aprovado' (>= 7.0), 'Recuperação' (>= 4.0) ou 'Reprovado' (< 4.0).

```python
# EXERCÍCIO 1
def classifica_nota(nota: float) -> str:
    if nota >= 7.0:
        return "Aprovado"
    elif nota >= 4.0:
        return "Recuperação"
    else:
        return "Reprovado"

# Teste de assertiva (não altere)
assert classifica_nota(7.5) == "Aprovado"
assert classifica_nota(5.0) == "Recuperação"
assert classifica_nota(3.9) == "Reprovado"
print("Função de classificação OK.")
```
