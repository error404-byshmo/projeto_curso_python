## Módulo 5: Estruturas de Repetição (`for` e `while`)

### 1. Loop `for` e `range()`
O loop `for` é ideal para iterar sobre sequências (como listas ou strings) ou para repetir ações um número conhecido de vezes (usando a função `range`).

A função `range(fim)` começa em 0. `range(início, fim)` começa no início. **O valor de `fim` nunca é incluído.**

```python
# Exemplo: Loop for com lista
cores = ["vermelho", "verde", "azul"]
for cor in cores:
    print(f"Cor atual: {cor}")
```

```python
# Exemplo: Loop for com range(início, fim)
for i in range(5, 8):
    print(i) # Imprime 5, 6, 7
```

### 2. Loop `while`
O loop `while` repete um bloco de código **enquanto** uma condição for avaliada como `True`. É importante que a condição mude dentro do loop para evitar um loop infinito.

```python
# Exemplo: Contador while
cont = 0
while cont < 4:
    print(f"Contando: {cont}")
    cont += 1 # Garante que a condição eventually será False
```

### 3. Exercícios Práticos
#### Exercício 1: Quadrados com Loop `for`
Use um loop `for` com `range()` para calcular e imprimir o quadrado dos números de 1 a 5.

```python
# EXERCÍCIO 1
quadrados = []
for num in range(1, 6):
    quadrado = num ** 2
    quadrados.append(quadrado)

# Teste de assertiva (não altere)
assert quadrados == [1, 4, 9, 16, 25]
print(f"Quadrados: {quadrados}")
```

#### Exercício 2: Divisores de um Número (`while`)
Encontre todos os divisores do número 12, começando a testar do 1 até o próprio número, usando um loop `while`.

```python
# EXERCÍCIO 2
numero = 12
divisor = 1
divisores = []

while divisor <= numero:
    if numero % divisor == 0:
        divisores.append(divisor)
    divisor += 1

# Teste de assertiva (não altere)
assert divisores == [1, 2, 3, 4, 6, 12]
print(f"Divisores de 12: {divisores}")
```
