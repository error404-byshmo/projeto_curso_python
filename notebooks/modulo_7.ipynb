## Módulo 7: Manipulação de Arquivos e Listas

### 1. Listas: Coleções Ordenadas
Listas são sequências de itens, definidas por colchetes `[]`. Os itens podem ser de diferentes tipos, e a contagem começa sempre do índice **0**.

```python
# Exemplo de Lista e Acesso por Índice
numeros = [10, 20, 30, 40]
print(f"O item no índice 2 é: {numeros[2]}")

# Fatiamento (Slicing): pega uma parte da lista (o final é não-inclusivo)
parte = numeros[1:3]
print(f"Fatiamento: {parte}")
```

### 2. Manipulação de Arquivos
Usamos a função `open()` em conjunto com a declaração `with` para garantir que os arquivos sejam fechados após a operação.

Modos:
* `'r'`: Leitura (read)
* `'w'`: Escrita (write - sobrescreve o conteúdo)
* `'a'`: Anexar (append - adiciona ao final)

```python
# Exemplo: Escrever e Ler Arquivo
arquivo_nome = 'dados_temp.txt'

# 1. Escrever
with open(arquivo_nome, 'w') as f:
    f.write("Linha A\n")
    f.write("Linha B")

# 2. Ler
with open(arquivo_nome, 'r') as f:
    conteudo = f.read()

print("Conteúdo do arquivo:")
print(conteudo)
```

### 3. Exercícios Práticos
#### Exercício 1: Soma dos Positivos em Lista
Crie uma função `soma_positivos(valores)` que itera sobre a lista e retorna a soma apenas dos números maiores que zero.

```python
# EXERCÍCIO 1
def soma_positivos(valores: list[float]) -> float:
    soma = 0.0
    for valor in valores:
        if valor > 0:
            soma += valor
    return soma

# Teste de assertiva (não altere)
assert soma_positivos([-3.0, 4.5, 0.0, 2.2, -1.1]) == 6.7
print(f"Soma dos positivos: {soma_positivos([1, -1, 5])}")
```

#### Exercício 2: Escrever Lista em Arquivo
Escreva cada elemento da lista `frutas` em uma nova linha no arquivo `frutas.txt`.

```python
# EXERCÍCIO 2
frutas = ["maçã", "banana", "uva", "laranja"]
arquivo_frutas = 'frutas.txt'

with open(arquivo_frutas, 'w') as f:
    for fruta in frutas:
        f.write(fruta + "\n")

# Teste de assertiva (Leitura)
with open(arquivo_frutas, 'r') as f:
    conteudo_lido = f.readlines()
assert len(conteudo_lido) == 4
assert 'uva\n' in conteudo_lido
print("Conteúdo de lista escrito no arquivo com sucesso.")
```
