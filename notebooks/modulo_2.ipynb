## Módulo 2: Variáveis e Tipos Primitivos

### 1. Variáveis
Variáveis são nomes que você usa para armazenar dados na memória. A atribuição é feita usando o sinal de igual (`=`).

```python
# Exemplo de criação de variável
contador = 7
pi_aproximado = 3.14
print(contador)
```

### 2. Tipos Primitivos de Dados
Python possui tipos de dados básicos (primitivos):
* **`int` (Inteiro):** Números sem casas decimais (e.g., 10, -5).
* **`float` (Real/Flutuante):** Números com casas decimais (e.g., 1.68, 9.99). **Use o ponto `.` como separador decimal.**
* **`bool` (Booleano):** Valores lógicos, apenas `True` ou `False`.

```python
# Exemplo de Booleanos
aprendendo_python = True
e_maior_de_idade = False
print(aprendendo_python)
```
