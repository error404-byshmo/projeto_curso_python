## Módulo 3: Strings e Operações Matemáticas

### 1. Operações Aritméticas
Python suporta as operações básicas:
* `+`, `-`, `*`, `/` (Adição, Subtração, Multiplicação, Divisão)
* `**` (Potência/Exponenciação)
* `//` (Divisão Inteira) e `%` (Módulo/Resto da divisão)

```python
# Exemplo de Módulo e Exponenciação
total_dias = 80
horas_em_um_dia = 24

dias = total_dias // horas_em_um_dia # Divisão Inteira
horas_restantes = total_dias % horas_em_um_dia # Resto

print(f"Dias: {dias}, Horas restantes: {horas_restantes}")
```

### 2. Strings e Métodos
Strings são sequências de caracteres. Elas possuem métodos úteis:
* `len()`: Obtém o tamanho da string.
* `.upper()` / `.lower()`: Converte para maiúsculas/minúsculas.

```python
# Exemplo de uso de métodos
palavra = "Programação"
print(f"Tamanho: {len(palavra)}")
print(f"Maiúsculas: {palavra.upper()}")
```

### 3. Entrada e Saída de Dados
* `print()`: Mostra informações na tela.
* `input()`: Solicita dados do usuário. O retorno é **sempre** uma string, que precisa ser convertida para número (`int()` ou `float()`) se for usada em cálculos.

```python
# Exemplo de input e conversão
entrada = input("Digite um número: ")
numero = float(entrada) # Converte a string para float
print(f"O dobro é: {numero * 2}")
```

### 4. Exercícios Práticos
#### Exercício 1: Média Ponderada
Calcule a média ponderada de duas notas e seus pesos:
* Nota 1: 9.0, Peso: 2
* Nota 2: 7.0, Peso: 3

```python
# EXERCÍCIO 1
n1, p1 = 9.0, 2
n2, p2 = 7.0, 3

media_ponderada = (n1 * p1 + n2 * p2) / (p1 + p2)

# Teste de assertiva (não altere)
assert round(media_ponderada, 2) == 7.80
print(f"Média Ponderada: {media_ponderada:.2f}")
```

#### Exercício 2: Volume da Esfera
Calcule o volume de uma esfera com raio $r=4$. A fórmula é $V = \frac{4}{3} \cdot \pi \cdot r^{3}$.
Lembre-se de importar `math.pi`.

```python
# EXERCÍCIO 2
import math

raio = 4
volume = (4/3) * math.pi * raio ** 3

# Teste de assertiva (não altere)
assert round(volume, 2) == 268.08
print(f"Volume da Esfera: {volume:.2f}")
```
