## Módulo 4: Estruturas Condicionais (`if`, `elif`, `else`)

### 1. O Fluxo de Controle
Condicionais permitem que seu programa tome decisões e siga diferentes caminhos de execução com base em condições lógicas.

```python
# Exemplo: if / else
saldo = 1500
valor_saque = 2000

if valor_saque <= saldo:
    print("Saque realizado com sucesso.")
    saldo -= valor_saque
else:
    print("Saldo insuficiente!")
```

### 2. Múltiplas Condições: `elif` e Operadores Lógicos
* `elif`: Usado para verificar uma condição secundária se o `if` principal for falso.
* `and`, `or`, `not`: Operadores que combinam condições.

```python
# Exemplo: Classificação de Idade
idade = 17

if idade >= 18:
    print("Adulto.")
elif idade >= 13 and idade < 18:
    print("Adolescente.")
else:
    print("Criança.")
```

### 3. Exercícios Práticos
#### Exercício 1: Positivo, Negativo ou Zero
Receba um número e determine se ele é positivo, negativo ou zero.

```python
# EXERCÍCIO 1
n = float(input("Digite um número: "))

status = ""
if n > 0:
    status = "Positivo"
elif n < 0:
    status = "Negativo"
else:
    status = "Zero"

# Teste de assertiva (Não altere)
if n == 5:
    assert status == "Positivo"
elif n == -2:
    assert status == "Negativo"
elif n == 0:
    assert status == "Zero"
print(f"O número é: {status}")
```

#### Exercício 2: Maior de Três Números
Receba três números diferentes (`a`, `b`, `c`) e encontre qual é o maior deles usando `if`/`elif`/`else`.

```python
# EXERCÍCIO 2
a = 12
b = 8
c = 15

maior_num = a
if b > maior_num:
    maior_num = b
if c > maior_num:
    maior_num = c

# Teste de assertiva (não altere)
assert maior_num == 15
print(f"O maior número é: {maior_num}")
```
