## Módulo 8: Introdução à Biblioteca Pandas

### 1. O que é Pandas?
Pandas é a biblioteca fundamental em Python para manipulação e análise de dados. Ela introduz as estruturas de dados **Series** (coluna) e **DataFrame** (tabela).

```python
# Importação padrão
import pandas as pd
print("Pandas importado como pd.")
```

### 2. Criando um DataFrame
Um DataFrame é criado a partir de um dicionário, onde as chaves são os nomes das colunas e os valores são listas de dados.

```python
# Exemplo de DataFrame
dados = {'Nome': ['Sara', 'Davi', 'Julia'],
         'Nota': [9.5, 7.0, 6.8],
         'Aprovado': [True, True, False]}
df = pd.DataFrame(dados)
print(df)
```

### 3. Análise e Filtragem
Podemos rapidamente obter estatísticas descritivas com `.describe()` e filtrar dados usando lógica booleana.

```python
# Exemplo: Resumo Estatístico e Filtragem
print("Estatísticas da coluna Nota:")
print(df['Nota'].describe())

print("\nAlunos Aprovados:")
df_aprovados = df[df['Nota'] >= 7.0]
print(df_aprovados)
```

### 4. Exercícios Práticos
#### Exercício 1: Criar e Resumir DataFrame
Crie um DataFrame `df_vendas` com as colunas 'Produto' e 'Quantidade' (com pelo menos 5 linhas). Em seguida, exiba o resumo estatístico da coluna 'Quantidade'.

```python
# EXERCÍCIO 1
dados_vendas = {'Produto': ['A', 'B', 'C', 'D', 'E'],
                'Quantidade': [15, 22, 5, 30, 18]}
df_vendas = pd.DataFrame(dados_vendas)

resumo_quantidade = df_vendas['Quantidade'].describe()

# Teste de assertiva (não altere)
assert len(df_vendas) == 5
assert 'mean' in resumo_quantidade
print("Resumo da Quantidade:")
print(resumo_quantidade)
```

#### Exercício 2: Filtrar Vendas Baixas
Filtre o `df_vendas` (do exercício anterior) para mostrar apenas os produtos com 'Quantidade' menor que 20.

```python
# EXERCÍCIO 2
df_vendas_baixas = df_vendas[df_vendas['Quantidade'] < 20]

# Teste de assertiva (não altere)
assert len(df_vendas_baixas) == 3
assert 'C' in df_vendas_baixas['Produto'].values
print("Produtos com baixa venda:")
print(df_vendas_baixas)
```
